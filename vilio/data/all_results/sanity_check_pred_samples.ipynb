```python
os.listdir(os.path.join(os.getcwd(), 'res'))
```

```python
keep_cols = ['id', 'proba_x', 'label_x']
col_map = {'proba_x': 'proba', 'label_x': 'label'}
e_folders = ['ES36', 'ESVCR36', 'ESV50', 'ES72', 'ESVCR72', 'EL36', 'ELVCR36', 'ELV50', 'EL72', 'ELVCR72']
```

```python
for ef in e_folders:
    print('-'*10 , ef, '-'*10)
    df = pd.read_csv(os.path.join('.', 'res', ef, ef+'_train.csv'))
    print(df.shape)
    if ef[:2] == 'ES':
        os.makedirs(os.path.join('.', 'res', 'ES_RECT', ef), exist_ok=True)
        for eff in os.listdir(os.path.join('.', 'res', ef)):            
            src = os.path.join('.', 'res', ef, eff)
            dest = os.path.join('.', 'res', 'ES_RECT', ef, eff)
            copyfile(src, dest)
            
            if 'train' in src:
                df = pd.read_csv(src)
                df_new = pd.merge(df, ref, on=['id'], how='inner')[keep_cols]
                df_new = df_new.rename(columns=col_map)
                df_new.to_csv(dest)
            
    elif ef[:2] == 'EL':
        os.makedirs(os.path.join('.', 'res', 'EL_RECT', ef), exist_ok=True)
        for eff in os.listdir(os.path.join('.', 'res', ef)):
            src = os.path.join('.', 'res', ef, eff)
            dest = os.path.join('.', 'res', 'EL_RECT', ef, eff)
            copyfile(src, dest)
            
            if 'train' in src:
                df = pd.read_csv(src)
                df_new = pd.merge(df, ref, on=['id'], how='inner')[keep_cols]
                df_new = df_new.rename(columns=col_map)
                df_new.to_csv(dest)
```

```python
for ef in os.listdir(os.path.join('.', 'res')):
    os.makedirs(os.path.join('.', 'newres', ef), exist_ok=True)
    for file in os.listdir(os.path.join('.', 'res', ef)):
        if 'dev' in file or 'test' in file:
            print(os.path.join('.', 'res', ef, file))
            df = pd.read_csv(os.path.join('.', 'res', ef, file)).sort_values(by='id')
        elif 'train' in file:
            print(os.path.join('.', 'res', ef, file))
            df = pd.read_csv(os.path.join('.', 'res', ef, file))
            delta = ref[~ref.id.isin(df.id)]
            df = df.append(delta).sort_values(by='id')
        df.to_csv(os.path.join('.', 'newres', ef, file))
```
